```python
# 4 Create a new categorical variable with value as Open and Closed. Open & Pending is to be categorized as Open and Closed & Solved is to be categorized as Closed.
```

```python
df['Status'].unique()
```

```python
df['new_status']=['Open' if st=="Open" or st=="pending" else "closed"  for st in df['Status']]
```

```python
df
```

```python
# 5 Provide state wise status of complaints in a stacked bar chart. Use the categorized variable from Q3. Provide insights on:
```

```python
state_complain=df.groupby(['State','new_status']).size().unstack().fillna(0)
```

```python
state_complain
```

```python
state_complain.plot.bar(stacked=True,figsize=(10,10))
```

```python
# 7 Which state has the highest percentage of unresolved complaint
```

```python
unresolved_data=df.groupby(['State','new_status']).size().unstack().fillna(0).sort_values(by='Open',ascending=False)
```

```python
unresolved_data
```

```python
unresolved_data['unresolved_cmp_percentage']=unresolved_data['Open']/unresolved_data['Open'].sum()*100
```

```python
unresolved_data
```

```python
# 8 Provide the percentage of complaints resolved till date, which were received through the Internet and customer care calls.
```

```python
resolved_data=df.groupby(['Received Via','new_status']).size().unstack()
```

```python
resolved_data
```

```python
resolved_data['resolved']=resolved_data['closed']/resolved_data['closed'].sum()*100
```
